```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px 
import seaborn as sns
```

Loading the dataset

```python
df = pd.read_csv("Unemployment in India.csv")
```

Inspecting the dataset and making the initial observations.

```python
df.info()
df.head()
```

The column names have hidden spaces (e.g., " Date"). We strip them

```python
df.columns = df.columns.str.strip()
```

Next, we drop all rows with missing values.

```python
df.dropna(inplace = True)
```

One of the most important steps in cleaning data with a Date column is ensuring that the date values are in the correct. This is my next step in my cleaning process. 

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst = True)
```

Next I will rename the column heads for easier use.

```python
df.head()
```

```python
df.rename(columns = {
    'Estimated Unemployment Rate (%)' : 'Unemployment_Rate',
    'Estimated Employed' : 'Employed',
    'Estimated Labour Participation Rate (%)' : 'Participation_Rate' }
          , inplace = True
         )
```

```python
df.head()
```

Now that the dataset is clean and ready for EDA. I am going to save it as a CSV file for easier export.

```python
df.to_csv("unemployment_data_India_cleaned.csv", index = False)
```
